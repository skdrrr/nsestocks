```python
import pandas as pd
from nsepy import get_history
from datetime import date
import time
from nsetools import Nse
nse = Nse()
print(nse)


NotAvailable = pd.DataFrame()
# List of Stocks
NIFTY200_stocks = pd.read_csv("/home/svdr/nsestocks/Universe/NIFTY200.csv")
NIFTY200_stocks = NIFTY200_stocks["Symbol"]

#Update data:

for element in NIFTY200_stocks:
    q = nse.get_quote(element)
    proxy_close = (q["buyPrice1"] + q["sellPrice1"])/2
    proxy_prev_close = q["previousClose"]
    proxy_open = q["open"]
    proxy_high = q["dayHigh"]
    proxy_low = q["dayLow"]
    proxy_last = q["lastPrice"]
    proxy_vwap = q["averagePrice"]
    proxy_volume = q["totalTradedVolume"]
    proxy_deliverable = q["deliveryQuantity"]
    proxy_perc_deliverable = q["deliveryToTradedQuantity"]/100
    proxy_turnover = q["averagePrice"] * q["totalTradedVolume"]

    intraday_row = {"Symbol":element, "Series":"EQ", "Prev Close":proxy_prev_close, "Open":proxy_open,"High":proxy_high,"Low":proxy_low,"Last":proxy_last,"Close":proxy_close,"VWAP":proxy_vwap,"Volume":proxy_volume,"Turnover":proxy_turnover,"Trades":0,"Deliverable Volume":proxy_deliverable,"%Deliverble":proxy_perc_deliverable}
    new_row = pd.DataFrame(intraday_row, index={date.today()})
    data = pd.read_pickle("/home/svdr/nsestocks/StocksRaw/"+element+".pickle")
    data = data.append(new_row)
    data.to_pickle("/home/svdr/nsestocks/intraday_data/"+element+".pickle")
    time.sleep(5)
    
    
```

```python
import pandas as pd
from nsepy import get_history
from datetime import date


NotAvailable = pd.DataFrame()
# List of Stocks
NSE_stocks = pd.read_csv("/home/svdr/nsestocks/Universe/NSEstocks.csv")
NIFTY200_stocks = pd.read_csv("/home/svdr/nsestocks/Universe/NIFTY200.csv")
NIFTY200_stocks = NIFTY200_stocks["Symbol"]

#Update data:
for element in NIFTY200_stocks:
    try:
        #read existing file
        current_data = pd.read_pickle("/home/svdr/nsestocks/StocksRaw/"+element+".pickle")
        last_day = current_data.index.max()
        #extract
        if last_day!=date.today():
            data = get_history(symbol=element, start=last_day, end=date.today())
            current_data = current_data.append(data)
            #remove duplicates
            current_data = current_data.loc[~current_data.index.duplicated(keep='first')]
            #save
            current_data.to_pickle("/home/svdr/nsestocks/StocksRaw/"+element+".pickle")
            print(element," done")
        else: 
            print(element,"already up to date")
    except:
        try:
            data = get_history(symbol=element, start=date(2008,1,1), end=date.today())
            data["Symbol"]=element
            data.to_pickle("/home/svdr/nsestocks/StocksRaw/"+element+".pickle")
        except:
            print(element,"NOT AVAILABLE")
            linie = {"Company":element}
            NotAvailable = NotAvailable.append(linie, ignore_index=True)

print("ALL DONE")
```
