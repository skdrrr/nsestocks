```python
## Take data from pickle and bundle it together to form close.pickle

import pandas as pd

# List of Stocks
NIFTY200_stocks = pd.read_csv("/home/svdr/nsestocks/Universe/NIFTY200.csv")
NIFTY200_stocks = NIFTY200_stocks["Symbol"]

#for first run, use:
close=pd.DataFrame()
# close = pd.read_pickle("/home/svdr/nsestocks/Data/close.pickle")
open_price=pd.DataFrame()
VWAP_price=pd.DataFrame()
high_price=pd.DataFrame()
low_price=pd.DataFrame()
volume=pd.DataFrame()
turnover=pd.DataFrame()
trades=pd.DataFrame()
deliverable=pd.DataFrame()


for element in NIFTY200_stocks:
    
    current_data = pd.read_pickle("/home/svdr/nsestocks/StocksRaw/"+element+".pickle")
    row_data = current_data[["Close"]].rename(columns={"Close":element})
    simbol = element
    close_data_c = current_data[["Close"]].rename(columns={'Close':simbol})
    open_price_c = current_data[["Open"]].rename(columns={'Open':simbol})
    VWAP_price_c = current_data[["VWAP"]].rename(columns={'VWAP':simbol})
    high_price_c = current_data[["High"]].rename(columns={'High':simbol})
    low_price_c = current_data[["Low"]].rename(columns={'Low':simbol})
    volume_c = current_data[["Volume"]].rename(columns={'Volume':simbol})
    turnover_c = current_data[["Turnover"]].rename(columns={'Turnover':simbol})
    trades_c = current_data[["Trades"]].rename(columns={'Trades':simbol})
    deliverable_c = current_data[["Deliverable Volume"]].rename(columns={'Deliverable Volume':simbol})
    
    if len(close)==0:
        close = close.append(close_data_c)
        open_price=open_price.append(open_price_c)
        VWAP_price=VWAP_price.append(VWAP_price_c)
        high_price=high_price.append(high_price_c)
        low_price=low_price.append(low_price_c)
        volume=volume.append(volume_c)
        turnover=turnover.append(turnover_c)
        trades=trades.append(trades_c)
        deliverable=deliverable.append(deliverable_c)
    else:
        try:
            close = close.join(close_data_c, how="outer")
            open_price=open_price.join(open_price_c, how="outer")
            VWAP_price=VWAP_price.join(VWAP_price_c, how="outer")
            high_price=high_price.join(high_price_c, how="outer")
            low_price=low_price.join(low_price_c, how="outer")
            volume=volume.join(volume_c, how="outer")
            turnover=turnover.join(turnover_c, how="outer")
            trades=trades.join(trades_c, how="outer")
            deliverable=deliverable.join(deliverable_c, how="outer")

            
            
        except:
            print("nu a functionat cu",element)
            
close.to_pickle("/home/svdr/nsestocks/Data/close.pickle")
open_price.to_pickle("/home/svdr/nsestocks/Data/open_price.pickle")
VWAP_price.to_pickle("/home/svdr/nsestocks/Data/VWAP_price.pickle")
high_price.to_pickle("/home/svdr/nsestocks/Data/high_price.pickle")
low_price.to_pickle("/home/svdr/nsestocks/Data/low_price.pickle")
volume.to_pickle("/home/svdr/nsestocks/Data/volume.pickle")
turnover.to_pickle("/home/svdr/nsestocks/Data/turnover.pickle")
trades.to_pickle("/home/svdr/nsestocks/Data/trades.pickle")
deliverable.to_pickle("/home/svdr/nsestocks/Data/deliverable.pickle")
```
